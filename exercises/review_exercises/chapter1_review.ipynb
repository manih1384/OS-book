### Chapter 1: Introduction

#### Section 1.1 What Operating Systems Do
**Review Question 1.1**
*   **Question:** What are the four components of a computer system?
*   **Answer:** Hardware, operating system, application programs, and users.

**Review Question 1.2**
*   **Question:** Provide at least three resources the operating system allocates.
*   **Answer:** CPU time, memory space, file storage space, and I/O devices.

**Review Question 1.3**
*   **Question:** What is the common name used to refer to the operating system program?
*   **Answer:** The kernel.

**Review Question 1.4**
*   **Question:** What do mobile operating systems often include in addition to the core kernel?
*   **Answer:** Middleware frameworks that support applications such as databases.

#### Section 1.2 Computer-System Organization
**Review Question 1.5**
*   **Question:** What is an interrupt?
*   **Answer:** An interrupt is a signal to the CPU emitted by hardware or software indicating an event that needs immediate attention.

**Review Question 1.6**
*   **Question:** What special operation triggers a software interrupt?
*   **Answer:** A system call (or monitor call).

**Review Question 1.7**
*   **Question:** What is one advantage of using a solid state disk over a magnetic disk?
*   **Answer:** Solid state disks are faster.

**Review Question 1.8**
*   **Question:** What is the difference between volatile and nonvolatile storage?
*   **Answer:** Volatile storage loses its contents when power is removed, while nonvolatile storage does not.

#### Section 1.3 Computer-System Architecture
**Review Question 1.9**
*   **Question:** What is another term for a multiprocessor system?
*   **Answer:** A parallel system or a multicore system.

**Review Question 1.10**
*   **Question:** Provide at least two advantages of multiprocessor systems.
*   **Answer:** Increased throughput, economy of scale, and increased reliability.

**Review Question 1.11**
*   **Question:** True or False? The most common multiple-processor system uses asymmetric multiprocessing.
*   **Answer:** False. The most common system uses symmetric multiprocessing (SMP).

**Review Question 1.12**
*   **Question:** What is the name of a multiprocessor system that uses multiple computing cores?
*   **Answer:** A multicore system.

**Review Question 1.13**
*   **Question:** How does a clustered system differ from a multicore system?
*   **Answer:** A clustered system gathers together multiple individual computer systems (nodes), while a multicore system has multiple computing cores on the same chip.

#### Section 1.4 Operating-System Structure
**Review Question 1.14**
*   **Question:** How does multiprogramming increase CPU utilization?
*   **Answer:** It organizes jobs so the CPU always has one to execute. When one job has to wait (e.g., for I/O), the operating system switches to another job.

**Review Question 1.15**
*   **Question:** What is the term for a program that has been loaded and is executing?
*   **Answer:** A process.

**Review Question 1.16**
*   **Question:** What part of the operating system makes the decision with regards to which job will run?
*   **Answer:** The CPU scheduler.

#### Section 1.5 Operating-System Operations
**Review Question 1.17**
*   **Question:** What are the two separate modes of operation?
*   **Answer:** User mode and kernel mode (also called supervisor mode, system mode, or privileged mode).

**Review Question 1.18**
*   **Question:** What is the mode of the system at boot time?
*   **Answer:** Kernel mode.

**Review Question 1.19**
*   **Question:** What is the mode of the system when the operating system gains control?
*   **Answer:** Kernel mode.

**Review Question 1.20**
*   **Question:** What is the mode of the system when a user program is running?
*   **Answer:** User mode.

**Review Question 1.21**
*   **Question:** Name at least two activities the operating system is responsible for in connection with memory management.
*   **Answer:** Keeping track of which parts of memory are currently being used and by whom; deciding which processes and data to move into and out of memory.

**Review Question 1.22**
*   **Question:** True or False? Managing files is one of the most visible aspects of an operating system.
*   **Answer:** True.

**Review Question 1.23**
*   **Question:** Name at least two activities the operating system is responsible for in connection with file management.
*   **Answer:** Creating and deleting files and directories; mapping files onto secondary storage; backing up files.

**Review Question 1.24**
*   **Question:** Name at least two activities the operating system is responsible for in connection with disk management.
*   **Answer:** Free-space management; storage allocation; disk scheduling.

**Review Question 1.25**
*   **Question:** Of the following five forms of storage, rank them from fastest to slowest in terms of access time: (1) main memory, (2) magnetic disk, (3) registers, (4) solid state disk, (5) cache.
*   **Answer:** Registers, cache, main memory, solid state disk, magnetic disk.

#### Section 1.6 Security and Protection
**Review Question 1.26**
*   **Question:** What is the difference between protection and security?
*   **Answer:** Protection is concerned with controlling the access of processes or users to the resources defined by a computer system. Security is concerned with defending a system from external or internal attacks.

#### Section 1.9 Kernel Data Structures
**Review Question 1.27**
*   **Question:** List at least four common kernel data structures.
*   **Answer:** Lists, stacks, queues, trees, and hash maps.

**Review Question 1.28**
*   **Question:** True or False? A bitmap of length N can be used to represent the status of 2N items.
*   **Answer:** False. A bitmap of length N can represent the status of N items.

#### Section 1.10 Computing Environments
**Review Question 1.29**
*   **Question:** List at least five different types of computing environments.
*   **Answer:** Traditional computing, mobile computing, client-server computing, peer-to-peer computing, cloud computing, real-time embedded systems.

#### Section 1.11 Open-Source Operating Systems
**Review Question 1.30**
*   **Question:** Provide an example of an open source operating system.
*   **Answer:** Linux.
