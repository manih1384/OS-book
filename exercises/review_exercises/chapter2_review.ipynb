
### **Operating Systems Review Questions**

**2.1 Operating-System Services**

**Question:** List at least three operating system services that are useful to users.
<br> **Answer:** Three operating system services useful to users are: (1) User Interface, which allows users to interact with the system, (2) Program Execution, which loads and runs applications, and (3) File System Manipulation, which provides capabilities to read, write, create, and delete files.

**Question:** List at least three operating system functions that maintain efficient operation of the system.
<br> **Answer:** Three operating system functions that maintain efficient operation are: (1) Resource Allocation, which distributes CPU time, memory, and storage among various processes, (2) Accounting, which tracks system resource usage by different users and processes, and (3) Protection and Security, which ensures controlled access to resources and defends the system from external threats.

**2.2 User and Operating-System Interface**

**Question:** What are the two different approaches for providing a user interface?
<br> **Answer:** The two different approaches for providing a user interface are: (1) Command-Line Interface (CLI), which uses text commands for user input, and (2) Graphical User Interface (GUI), which uses visual elements like windows, icons, and menus for user interaction.

**2.3 System Calls**

**Question:** What is a system call?
<br> **Answer:** A system call is a programming interface that provides a controlled entry point into the operating system kernel, allowing a running application to request a service from the operating system, such as performing I/O operations or creating a new process.

**Question:** What is an API?
<br> **Answer:** An API (Application Programming Interface) is a set of defined functions, protocols, and tools for building application software. It often provides a convenient, high-level interface for programmers to use system calls without needing to know the lower-level implementation details.

**Question:** What kernel data structure can be used for one technique of passing parameters to system calls?
<br> **Answer:** A **block** or **table** in kernel memory can be used for one technique of passing parameters to system calls. Specifically, the address of this block is passed as a parameter in a register.

**2.4 System Services**

**Question:** List at least three of the major categories of system calls.
<br> **Answer:** Three major categories of system calls are: (1) Process Control (e.g., create, terminate a process), (2) File Management (e.g., create, read, write files), and (3) Device Management (e.g., request or release a device).

**Question:** A program that has been loaded and executing is called a \_\_\_\_\_\_\_\_.
<br> **Answer:** A program that has been loaded and executing is called a **process**.

**Question:** What part of the operating system makes the decision with regards to which job will run?
<br> **Answer:** The part of the operating system that makes the decision regarding which job will run is the **CPU scheduler**, which is a component of the kernel.

**2.7 Operating-System Design and Implementation**

**Question:** What are the two basic goal groups that must be considered when designing an operating system?
<br> **Answer:** The two basic goal groups for operating system design are: (1) User Goals, which focus on making the system convenient, easy to use, and reliable, and (2) System Goals, which focus on making the system efficient, stable, secure, and easy to implement and maintain.

**Question:** What is the difference between policy and mechanism?
<br> **Answer:** The difference between policy and mechanism is that a **mechanism** refers to *how* something is to be done (the implementation), while a **policy** refers to *what* is to be done (the decision-making rule). Mechanisms determine how resources are allocated, while policies decide which processes get those resources.


**2.8 Operating-System Structure**

**Question:** List at least three different ways for structuring an operating system.
<br> **Answer:** Three different ways for structuring an operating system are: (1) Simple (Monolithic) structure, (2) Layered approach, and (3) Microkernel architecture.

**Question:** List at least two different hybrid operating systems.
<br> **Answer:** Two different hybrid operating systems are macOS and Windows NT (and its successors, such as Windows 10 and 11). These systems combine aspects of monolithic and microkernel architectures.

**Question:** What are the two devices that run the iOS operating system?
<br> **Answer:** The two primary devices that run the iOS operating system are the iPhone and the iPad.

**Question:** What technique do microkernels use to communicate between services?
<br> **Answer:** Microkernels use a technique called **message passing** to facilitate communication between services (or processes) running in user space.

**Question:** Provide an example of an operating system that uses the simple structure.
<br> **Answer:** An example of an operating system that uses the simple (monolithic) structure is MS-DOS.

**2.10 Debugging and System Performance**

**Question:** True or False? Performance tuning is a type of debugging.
<br> **Answer:** **True.** Performance tuning is a form of debugging that focuses on improving performance rather than fixing correctness errors.

**Question:** True or False? DTrace is available for Windows systems.
<br> **Answer:** **False.** DTrace is a dynamic tracing framework originally developed for Solaris and is also available for macOS and FreeBSD, but it is not natively available for Windows systems.

**Question:** Name two activities the operating system is responsible for in connection with disk management.
<br> **Answer:** The operating system is responsible for (1) **Disk formatting**, which initializes the disk's logical structure, and (2) **Managing free space**, which involves tracking and allocating available storage blocks.

**Question:** Name at least two activities the operating system is responsible for in connection with disk management.
<br> **Answer:** Two activities the operating system is responsible for in connection with disk management are (1) **Handling I/O operations** for reading from and writing to the disk, and (2) **Providing a file system** to organize data into files and directories for easy storage and retrieval.

**Question:** Of the following 5 forms of storage, rank them from fastest to slowest in terms of access time: (1) main memory, (2) magnetic disk, (3) registers, (4) solid state disk, (5) cache.
<br> **Answer:** Ranked from fastest to slowest access time: **(3) Registers** -> **(5) Cache** -> **(1) Main Memory** -> **(4) Solid State Disk** -> **(2) Magnetic Disk**.
