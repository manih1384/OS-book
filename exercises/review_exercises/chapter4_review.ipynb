**Review Question 4.1**
*   **Question:** How many threads does a traditional, heavyweight process have?
*   **Answer:** One (a single thread of execution).

**Review Question 4.2**
*   **Question:** Provide at least three benefits of multithreaded programming.
*   **Answer:** 1) Responsiveness (allowing a program to continue running even if part of it is blocked), 2) Resource sharing (threads share memory and resources of the process), 3) Economy (creating and context-switching threads is cheaper than processes), 4) Scalability (can take advantage of multicore architectures).

**Review Question 4.3**
*   **Question:** True or False? Concurrency is only possible with parallelism.
*   **Answer:** False. Concurrency supports more than one task making progress (e.g., via time-sharing on a single core), while parallelism requires multiple tasks to execute simultaneously (on multiple cores/CPUs). Concurrency can exist without parallelism.

**Review Question 4.4**
*   **Question:** True or False? Amdahl’s Law addresses the disproportionate effect of the serial portion of a program.
*   **Answer:** True. Amdahl's Law states that the speedup gained from parallel processing is limited by the sequential (serial) portion of the program.

**Review Question 4.5**
*   **Question:** List at least three challenges when designing programming for multicore systems.
*   **Answer:** 1) Dividing activities (identifying tasks that can run in parallel), 2) Balance (ensuring tasks take roughly the same amount of time), 3) Data splitting (dividing data for concurrent access), 4) Data dependency (synchronizing tasks that depend on each other's data), 5) Testing and debugging (complexity increases with concurrency).

**Review Question 4.6**
*   **Question:** What are the two general types of parallelism?
*   **Answer:** Data parallelism and task parallelism.

**Review Question 4.7**
*   **Question:** List the three common ways of mapping user threads to kernel threads.
*   **Answer:** 1) Many-to-one model, 2) One-to-one model, 3) Many-to-many model.

**Review Question 4.8**
*   **Question:** True or False? Only Linux and Windows implement the one-to-one model.
*   **Answer:** False. While Linux and Windows use the one-to-one model, it is a common model and not exclusive to them (e.g., modern implementations of Solaris also support it).

**Review Question 4.9**
*   **Question:** What are the two approaches for implementing a thread library?
*   **Answer:** 1) A library entirely in user space, 2) A kernel-level library supported directly by the operating system.

**Review Question 4.10**
*   **Question:** What are the three main thread libraries in use?
*   **Answer:** POSIX Pthreads, Windows threads, and Java threads.

**Review Question 4.11**
*   **Question:** True or False? PThreads is typically only implemented on UNIX-like systems.
*   **Answer:** False. While Pthreads originated and is standard on UNIX-like systems (Linux, macOS, Solaris), implementations exist for other platforms, including Windows.

**Review Question 4.12**
*   **Question:** True or False? PThreads is only a specification, not an implementation.
*   **Answer:** True. POSIX Pthreads is an API specification defined by the IEEE POSIX standards. Different operating systems provide their own implementations of this specification.

**Review Question 4.13**
*   **Question:** What is the PThread API for creating a thread?
*   **Answer:** `pthread_create()`.

**Review Question 4.14**
*   **Question:** What is the Windows API for creating a thread?
*   **Answer:** `CreateThread()` (or the `_beginthreadex()` helper function from the C runtime library).


**Review Question 4.15**
*   **Question:** What Java method is used for allocating and initializing a new thread in the JVM?
*   **Answer:** The `start()` method of the `Thread` class.

**Review Question 4.16**
*   **Question:** Provide at least two techniques for supporting implicit threading.
*   **Answer:** Thread pools, fork-join (or fork-join parallelism), OpenMP, Grand Central Dispatch (GCD).

**Review Question 4.17**
*   **Question:** True or False? Grand Central Dispatch only works for Apple’s Mac OS X and iOS operating systems.
*   **Answer:** True. Grand Central Dispatch (GCD) is a technology developed by Apple for its operating systems (macOS, iOS, tvOS, watchOS).

**Review Question 4.18**
*   **Question:** True or False? The semantics of the `fork()` system call can vary on multithreaded systems.
*   **Answer:** True. The behavior of `fork()` in a multithreaded program is a significant issue. Some systems have two versions: one that duplicates all threads and one that duplicates only the calling thread. The POSIX standard specifies `fork` to create a process with a single thread (a copy of the thread that called `fork`).

**Review Question 4.19**
*   **Question:** What are the two scenarios for canceling a target thread?
*   **Answer:** 1) **Asynchronous cancellation:** One thread immediately terminates the target thread. 2) **Deferred cancellation:** The target thread periodically checks if it should terminate, allowing it to perform necessary cleanup.

**Review Question 4.20**
*   **Question:** What is the PThreads API for thread cancellation?
*   **Answer:** `pthread_cancel()` (to request cancellation of a thread).

**Review Question 4.21**
*   **Question:** True or False? Windows threads provide both user and kernel stacks.
*   **Answer:** True. Each Windows thread has both a user-mode stack and a kernel-mode stack.

**Review Question 4.22**
*   **Question:** What term does Linux use to refer to a process or a thread?
*   **Answer:** Task. In Linux, both traditional processes and threads are implemented as tasks (`task_struct` in the kernel). Threads are simply tasks that share the same address space and other resources.
