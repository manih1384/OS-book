### Chapter 8: Deadlocks

**Review Question 8.1**
*   **Question:** True or False? The system model for deadlocks first requires a process request a resource, then use the resource, and finally release the resource.
*   **Answer:** True. The standard resource usage model is: Request → Use → Release.

**Review Question 8.2**
*   **Question:** What are the four necessary conditions for characterizing deadlock?
*   **Answer:** 1) Mutual Exclusion, 2) Hold and Wait, 3) No Preemption, 4) Circular Wait.

**Review Question 8.3**
*   **Question:** Describe one strategy for dealing with deadlocks.
*   **Answer:** One strategy is **Deadlock Prevention**, which ensures that at least one of the four necessary conditions for deadlock cannot hold (e.g., requiring processes to request all resources before starting to break "hold and wait").

**Review Question 8.4**
*   **Question:** What is the only reasonable condition that can be used to prevent deadlocks from occurring?
*   **Answer:** The "Circular Wait" condition is generally considered the most practical to prevent, typically by imposing a total ordering of all resource types and requiring that processes request resources in an increasing order of enumeration.

**Review Question 8.5**
*   **Question:** What is the name of the state of the system if resources can be allocated to all processes in some order and deadlock can still be avoided?
*   **Answer:** Safe state.

**Review Question 8.6**
*   **Question:** What is the name of the classic deadlock avoidance algorithm?
*   **Answer:** The banker's algorithm.

**Review Question 8.7**
*   **Question:** True or False? The wait-for graph can only be used for deadlock detection when there is a single instance of each resource type.
*   **Answer:** True. The standard wait-for graph detection method applies to systems where each resource type has exactly one instance. For multiple instances, a more generalized algorithm (like the Banker's algorithm variant for detection) is used.

**Review Question 8.8**
*   **Question:** Provide at least one method for recovering from deadlock.
*   **Answer:** Process termination (aborting one or more deadlocked processes) or resource preemption (forcibly taking resources from some processes and giving them to others).
