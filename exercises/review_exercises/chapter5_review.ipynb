### Chapter 5: CPU Scheduling

**Review Question 5.1**
*   **Question:** What are the two bursts that CPU schedulers are designed around?
*   **Answer:** CPU bursts and I/O bursts.

**Review Question 5.2**
*   **Question:** True or False? Under preemptive scheduling, when a process switches from the running to the ready state, it may lose control of the CPU.
*   **Answer:** True. In preemptive scheduling, the scheduler can interrupt a running process (e.g., when a higher-priority process becomes ready or a time quantum expires) and reassign the CPU.

**Review Question 5.3**
*   **Question:** List at least three different criteria for designing a CPU scheduling algorithm.
*   **Answer:** CPU utilization, throughput, turnaround time, waiting time, response time.

**Review Question 5.4**
*   **Question:** What scheduling algorithm assigns the CPU to the process with the highest priority?
*   **Answer:** Priority scheduling.

**Review Question 5.5**
*   **Question:** True or False? The multilevel feedback queue scheduling algorithm allows processes to migrate between different queues.
*   **Answer:** True. A key feature of the Multilevel Feedback Queue (MLFQ) scheduler is that processes can move between queues based on their observed behavior (e.g., CPU-bound processes may be moved to lower-priority queues).

**Review Question 5.6**
*   **Question:** What scheduling algorithm assigns the CPU to the process that first requested it?
*   **Answer:** First-Come, First-Served (FCFS) scheduling.

**Review Question 5.7**
*   **Question:** What scheduling algorithm assigns the CPU to a process for only its time slice (or time quantum)?
*   **Answer:** Round-Robin (RR) scheduling.

**Review Question 5.8**
*   **Question:** What scheduling algorithm assigns the CPU to the process with the shortest burst?
*   **Answer:** Shortest-Job-First (SJF) scheduling (or Shortest-Process-Next, SPN).

**Review Question 5.9**
*   **Question:** What are the two types of contention scope for thread scheduling?
*   **Answer:** Process-contention scope (PCS) and system-contention scope (SCS).

**Review Question 5.10**
*   **Question:** What are the two general hardware instructions that can be performed atomically?
*   **Answer:** Test-and-set and compare-and-swap (or swap).

**Review Question 5.11**
*   **Question:** What is more common on current systems, asymmetric or symmetric multiprocessing?
*   **Answer:** Symmetric multiprocessing (SMP).

**Review Question 5.12**
*   **Question:** What are the two forms of processor affinity?
*   **Answer:** Soft affinity (the operating system tries to keep a process on the same processor but does not guarantee it) and hard affinity (allows a process to specify a set of processors on which it may run).

**Review Question 5.13**
*   **Question:** What are the two general approaches for load balancing?
*   **Answer:** Push migration (a specific task periodically checks the load on each processor and redistributes processes if needed) and pull migration (an idle processor pulls a waiting task from a busy processor).

**Review Question 5.14**
*   **Question:** What are the two ways to multithread a processing core?
*   **Answer:** Coarse-grained multithreading and fine-grained multithreading (also known as simultaneous multithreading, SMT, in its most common modern form, e.g., Intel's Hyper-Threading).

**Review Question 5.15**
*   **Question:** What are the two general types of real-time scheduling?
*   **Answer:** Soft real-time scheduling and hard real-time scheduling.

**Review Question 5.16**
*   **Question:** What real-time scheduling algorithm uses deadline as its scheduling criteria?
*   **Answer:** Earliest-Deadline-First (EDF) scheduling.

**Review Question 5.17**
*   **Question:** What real-time scheduling algorithm is used for scheduling periodic tasks with static priorities?
*   **Answer:** Rate-Monotonic Scheduling (RMS).

**Review Question 5.18**
*   **Question:** What is the name of the default scheduling algorithm for current Linux systems?
*   **Answer:** The Completely Fair Scheduler (CFS).

**Review Question 5.19**
*   **Question:** True or False? A Windows thread is assigned both a priority class and a relative priority within that class.
*   **Answer:** True. Each thread has a base priority determined by its process's priority class and its own relative priority within that class.

**Review Question 5.20**
*   **Question:** If a thread on a Solaris system exhausts its time quantum, will it later be assigned a higher or lower priority?
*   **Answer:** Lower priority. In the traditional Solaris time-sharing (TS) class, a thread that uses its entire time quantum is considered CPU-bound and has its priority lowered.

**Review Question 5.21**
*   **Question:** True or False? Deterministic modeling and simulations are similar strategies for evaluating scheduling algorithms.
*   **Answer:** False. They are different strategies. Deterministic modeling uses a predetermined workload (e.g., a fixed set of processes with known burst times) to compute performance metrics analytically. Simulation involves modeling a system (e.g., with software) and evaluating algorithm performance using more complex, often probabilistic, workloads.
