### Chapter 1: Introduction

#### Practice Exercises

**Practice Exercise 1.1**
*   **Question:** What are the three main purposes of an operating system?
*   **Answer:** The three main purposes are:
    *   To provide an environment for a computer user to execute programs on computer hardware in a convenient and efficient manner.
    *   To allocate the separate resources of the computer as needed to perform the required tasks. The allocation process should be as fair and efficient as possible.
    *   As a control program, it serves two major functions: (1) supervision of the execution of user programs to prevent errors and improper use of the computer, and (2) management of the operation and control of I/O devices.

**Practice Exercise 1.2**
*   **Question:** We have stressed the need for an operating system to make efficient use of the computing hardware. When is it appropriate for the operating system to forsake this principle and to “waste” resources? Why is such a system not really wasteful?
*   **Answer:** Single-user systems should maximize use of the system for the user. A GUI might “waste” CPU cycles, but it optimizes the user’s interaction with the system.

**Practice Exercise 1.3**
*   **Question:** What is the main difficulty that a programmer must overcome in writing an operating system for a real-time environment?
*   **Answer:** The main difficulty is keeping the operating system within the fixed time constraints of a real-time system. If the system does not complete a task in a certain time frame, it may cause a breakdown of the entire system. Therefore, when writing an operating system for a real-time system, the writer must be sure that his scheduling schemes don’t allow response time to exceed the time constraint.

**Practice Exercise 1.4**
*   **Question:** Keeping in mind the various definitions of operating system, consider whether the operating system should include applications such as web browsers and mail programs. Argue both that it should and that it should not, and support your answers.
*   **Answer:**
    *   **Argument for:** If the application is embedded within the operating system, it is likely to be better able to take advantage of features in the kernel and therefore have performance advantages over an application that runs outside of the kernel.
    *   **Argument against:** (1) the applications are applications—not part of an operating system, (2) any performance benefits of running within the kernel are offset by security vulnerabilities, and (3) inclusion of applications leads to a bloated operating system.

**Practice Exercise 1.5**
*   **Question:** How does the distinction between kernel mode and user mode function as a rudimentary form of protection (security)?
*   **Answer:** Certain instructions can be executed only when the CPU is in kernel mode. Similarly, hardware devices can be accessed only when the program is in kernel mode, and interrupts can be enabled or disabled only when the CPU is in kernel mode. Consequently, the CPU has very limited capability when executing in user mode, thereby enforcing protection of critical resources.

**Practice Exercise 1.9**
*   **Question:** Timers could be used to compute the current time. Provide a short description of how this could be accomplished.
*   **Answer:** A program could set a timer for some time in the future and go to sleep. When awakened by the interrupt, it could update its local state, which it uses to keep track of the number of interrupts it has received thus far. It could then repeat this process of continually setting timer interrupts and updating its local state when the interrupts are actually raised.

**Practice Exercise 1.10**
*   **Question:** Give two reasons why caches are useful. What problems do they solve? What problems do they cause? If a cache can be made as large as the device for which it is caching (for instance, a cache as large as a disk), why not make it that large and eliminate the device?
*   **Answer:**
    *   **Usefulness:** Caches are useful when two or more components need to exchange data, and the components perform transfers at differing speeds. Caches solve the transfer problem by providing a buffer of intermediate speed between the components.
    *   **Problems Solved:** If the fast device finds the data it needs in the cache, it need not wait for the slower device.
    *   **Problems Caused:** The data in the cache must be kept consistent with the data in the components. If a component has a data value change, and the datum is also in the cache, the cache must also be updated.
    *   **Why not replace the device?** A component may be eliminated by an equal-sized cache, but only if: (a) the cache and the component have equivalent state-saving capacity, and (b) the cache is affordable, because faster storage tends to be more expensive.

**Practice Exercise 1.11**
*   **Question:** Distinguish between the client–server and peer-to-peer models of distributed systems.
*   **Answer:** The client-server model firmly distinguishes the roles of the client and server. Under this model, the client requests services that are provided by the server. The peer-to-peer model doesn’t have such strict roles. In fact, all nodes in the system are considered peers and thus may act as either clients or servers—or both. A node may request a service from another peer, or the node may in fact provide such a service to other peers in the system.

**Exercise 1.12**
*   **Question:** How do clustered systems differ from multiprocessor systems? What is required for two machines belonging to a cluster to cooperate to provide a highly available service?
*   **Answer:** A multiprocessor system has multiple processors within a single machine sharing main memory, while a clustered system couples multiple independent machines (nodes) via a network. For two machines to cooperate for high availability, a cluster requires a layer of cluster software that provides services like a heartbeat to monitor node status and shared storage for data.

**Exercise 1.13**
*   **Question:** Consider a computing cluster consisting of two nodes running a database. Describe two ways in which the cluster software can manage access to the data on the disk. Discuss the benefits and disadvantages of each.
*   **Answer:**
    1.  **Asymmetric Clustering:** One machine runs the database application while the other is in hot-standby, only monitoring the first.
        *   **Benefit:** Simple to implement.
        *   **Disadvantage:** The standby server is idle, wasting resources.
    2.  **Symmetric Clustering:** Multiple machines run the database application and manage access to shared storage concurrently.
        *   **Benefit:** More efficient use of hardware resources.
        *   **Disadvantage:** Requires more sophisticated software to manage concurrent access and avoid data corruption.

**Exercise 1.14**
*   **Question:** What is the purpose of interrupts? How does an interrupt differ from a trap? Can traps be generated intentionally by a user program? If so, for what purpose?
*   **Answer:** The purpose of interrupts is to alert the CPU to high-priority events that require attention, allowing it to defer normal processing. An interrupt is an asynchronous event triggered by a signal from an external device, while a trap is a synchronous event generated by an error or a specific instruction in the executing program. Yes, traps can be generated intentionally by a user program. The purpose is to execute a system call, requesting a service from the operating system.

**Exercise 1.15**
*   **Question:** Explain how the Linux kernel variables HZ and jiffies can be used to determine the number of seconds the system has been running since it was booted.
*   **Answer:** The variable `HZ` is the frequency of the system timer interrupt in hertz (interrupts per second). The variable `jiffies` counts the number of timer interrupts that have occurred since boot. The system uptime in seconds is calculated as `jiffies / HZ`.

**Exercise 1.16**
*   **Question:** Direct memory access is used for high-speed I/O devices in order to avoid increasing the CPU’s execution load.
    a. How does the CPU interface with the device to coordinate the transfer?
    b. How does the CPU know when the memory operations are complete?
    c. The CPU is allowed to execute other programs while the DMA controller is transferring data. Does this process interfere with the execution of the user programs? If so, describe what forms of interference are caused.
*   **Answer:**
    a. The CPU programs the DMA controller by setting registers to specify the source/destination address, the transfer count, and the direction of the transfer. It then issues a command to the device to start the transfer.
    b. The DMA controller sends an interrupt signal to the CPU once the entire data transfer is complete.
    c. Yes, it can interfere. While the DMA controller does not use the CPU's execution cycles, it does compete for memory cycles by accessing memory directly. This contention for the memory bus can slow down CPU memory accesses, slowing the execution of user programs.

**Exercise 1.17**
*   **Question:** Some computer systems do not provide a privileged mode of operation in hardware. Is it possible to construct a secure operating system for these computer systems? Give arguments both that it is and that it is not possible.
*   **Answer:**
    *   **Argument for it being possible:** It is possible if every program is written to be well-behaved and voluntarily does not interfere with other programs or the operating system. Security could be enforced through software interpretation (e.g., a Java virtual machine) that checks all instructions.
    *   **Argument for it being not possible:** It is not possible to construct a truly secure system. Without hardware support for privilege separation, a user program could accidentally or maliciously issue instructions that corrupt the operating system or other user programs, and there would be no hardware mechanism to prevent it.

**Exercise 1.18**
*   **Question:** Many SMP systems have different levels of caches; one level is local to each processing core, and another level is shared among all processing cores. Why are caching systems designed this way?
*   **Answer:** This design provides a balance between speed and data sharing. Local caches (L1/L2) are very fast and reduce latency for a core's private data. A shared cache (L3) is larger and allows efficient sharing of data between cores, while still being faster than main memory.

**Exercise 1.21**
*   **Question:** Discuss, with examples, how the problem of maintaining coherence of cached data manifests itself in the following processing environments:
    a. Single-processor systems
    b. Multiprocessor systems
    c. Distributed systems
*   **Answer:**
    a. **Single-processor systems:** Coherence is between the CPU cache and main memory. For example, after a DMA transfer from a disk controller directly to memory, the CPU's cache may hold stale data and must be invalidated.
    b. **Multiprocessor systems:** Coherence is between the caches of multiple processors. For example, if one processor updates a shared variable in its cache, other processors must be made aware of this change to prevent using outdated values.
    c. **Distributed systems:** Coherence is between the entire memory of multiple machines. For example, replicating a file across several servers requires a protocol to ensure all copies are updated when a change is made to one, maintaining a consistent view for all clients.

**Exercise 1.22**
*   **Question:** Describe a mechanism for enforcing memory protection in order to prevent a program from modifying the memory associated with other programs.
*   **Answer:** A base and limit register pair can be used. The base register holds the smallest legal physical memory address for the program, and the limit register specifies the size of the range. The CPU checks every memory address generated by the program to ensure it is between the base and base+limit, preventing access to memory outside its allocated region.
